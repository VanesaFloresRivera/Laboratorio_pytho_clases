````python
#1.1. Define  la clase Cancion, la cual va a tener 4 atributos básicos. (título,  artista, album, genero y duración):
class Cancion:
    def __init__(self, titulo, artista, album, duracion, genero):
        self.titulo = titulo
        self.artista = artista
        self.album = album
        self.duracion = duracion
        self.genero = genero
    
 #1.2. Crea un método para la clase Cancion que nos muestre la información de la canción.   
    def informacion_cancion (self):
        self.__dict__
        return self.__dict__
#1.3. Crea un método que nos permita comparar la duración de dos canciones y nos indique cuál de las dos es más larga.   
    def cancion_mas_larga (self, cancion2):
        if self.duracion> cancion2.duracion:
            print(f'La duración de la canción {self.titulo} es {self.duracion}')
            print(f'La duración de la canción {cancion2.titulo} es {cancion2.duracion}')
            return f'La cancion {self.titulo} es la más larga'
        else:
            print(f'La duración de la canción {self.titulo} es {self.duracion}')
            print(f'La duración de la canción {cancion2.titulo} es {cancion2.duracion}')
            return f'La cancion {cancion2.titulo} es la más larga'
#1.4. Crea un método que nos permita calcular la popularidad de una canción en función de varios criterios.

    def popularidad (self, album_perteneciente):
        #- La puntuación base debe ser de 50.
        puntuacion_base = 50
        #- Las canciones de entre 2 y 4 minutos de duración son más populares, por lo que suman 20 puntos. Las demás restan 10.  
        if self.duracion > 2 and self.duracion<4:
            puntuacion_base +=20
        else:
            puntuacion_base -=10
        #- Las canciones de los géneros pop, rock, hip-hop y electrónicas son populares, por lo que suman 20 puntos. El resto de géneros restan 5 puntos.
        if self.genero.lower() == 'pop' or self.genero.lower() == 'rock' or self.genero.lower() == 'hip-hop' or self.genero.lower() == 'electronicas':
            puntuacion_base +=20
        else:
            puntuacion_base -=5
        #-Si una canción pertenece a un álbum suma 10 puntos.
        if self.album == album_perteneciente:
            puntuacion_base +=10
        else:
            puntuacion_base = puntuacion_base
        print(f'La canción {self.titulo} tiene una popularidad de:')
        return puntuacion_base
    
````

````python
#4.1. Define  la clase Playlist, la cual va a tener 1 atributo básico. (nombre) y va a tener una lista con las canciones (inicialmente estará vacía)

class Playlist:
    def __init__(self, nombre, lista_canciones =[]):
        self.nombre = nombre
        self.lista_canciones = lista_canciones

#4.2. Crea un método para la clase Playlist que nos permita añadir una canción a nuestra Playlist. (Esta canción va a ser un objeto de la clase Cancion)
    def añadir_cancion(self, canciones_añadir):
        print(f'La longitud de la lista de canciones antes de la actualización es {len(self.lista_canciones)}')
        for cancion in canciones_añadir:
            #print(cancion)
            #print(len(canciones))
            #print(str(cancion[0]))
            #cancion_obj = clase(cancion[0], cancion[1], cancion[2], cancion[3], cancion[4])
            #print(cancion_obj.__dict__)
            self.lista_canciones.append(cancion)
        print(f'La longitud de la lista de canciones tras la actualización es {len(self.lista_canciones)}')
        print(f'La lista de canciones actualizadas es:')
        return self.lista_canciones

#4.3. Crea un método que nos permita eliminar canciones de nuestra Playlist.
    def eliminar_cancion(self, canciones_eliminar):
        print(f'La longitud de la lista de canciones antes de la actualización es {len(self.lista_canciones)}')
        for cancion in canciones_eliminar:
            #print(cancion)
            #print(self.lista_canciones)
            self.lista_canciones.remove(cancion)
        print(f'La longitud de la lista de canciones tras la actualización es {len(self.lista_canciones)}')
        print(f'La lista de canciones actualizadas es:')
        return self.lista_canciones

#4.4. Crea un método que nos permita visualizar el nombre de nuestra playlist, las canciones que tiene y su información. 
    def informacion_playlist (self):
        print(self.__dict__)
        return(help(self))
    
#4.5. Crea un método que nos permita encontrar y visualizar todas las canciones de un artista concreto de nuestra Playlist
    def canciones_totales_artista (self, artista_concreto):
        canciones_artista=[]
        titulos_canciones = []
        for cancion in self.lista_canciones:
            #print(cancion.__dict__)
            artista_canción = cancion.__dict__.get('artista')
            #print(artista_canción)
            #print(diccionario_canción.get('artista'))
            if artista_canción == artista_concreto:
                canciones_artista.append(cancion)
                titulo_cancion= cancion.__dict__.get('titulo')
                titulos_canciones.append(titulo_cancion)
            else:
                continue
        print(f'Las canciones totales de {artista_concreto} son:\n {canciones_artista} \n y sus títulos son \n {titulos_canciones}')

#4.6. Crea un método para contar el número total de canciones de nuestra playlist y su duración total
    def canciones_totales_duración_total (self):
        print(f'El número total de canciones es {len(self.lista_canciones)}')
        duración_total = 0
        for cancion in self.lista_canciones:
            duración_canción = cancion.__dict__.get('duracion') 
            duración_total += duración_canción
        print(f'La duración total de las canciones es: ')
        return duración_total
    
#4.7. Crea un método que nos permita ordenar aleatoriamente las canciones de nuestra playlist
    def orden_aleatorio (self):
        import random
        random.shuffle(self.lista_canciones)
        return self.lista_canciones
        
````
